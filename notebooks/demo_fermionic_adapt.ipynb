```python
from openvqe.vqe import VQE
```

```python
molecule_symbol = 'H2'
type_of_generator = 'spin_complement_gsd'
transform = 'JW'

opts = {
        'n_max_grads': 1,
        'optimizer': 'COBYLA',
        'tolerance': 10**(-6),
        'type_conver': 'norm',
        'threshold_needed': 1e-2,
        'max_external_iterations': 35
    }
```

```python
VQE('fermionic_adapt', molecule_symbol, type_of_generator, transform, False, opts).execute()
```

```python
VQE('fermionic_adapt', molecule_symbol, type_of_generator, transform, True, opts).execute()
```
